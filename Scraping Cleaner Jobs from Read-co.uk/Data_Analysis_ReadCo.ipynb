```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
readcouk_data = pd.read_csv('Read-co-uk_Portal.csv',index_col='Unnamed: 0')
readcouk_data.head()
```

```python
readcouk_data['Area/Zip'] = readcouk_data['Area/Zip'].str.replace(" \n London"," ")
```

```python
readcouk_data['Area/Zip'].value_counts().head(10)
```

```python
readcouk_data['Role'].value_counts().head(10)
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_autopct
```

```python
values_by_Area = readcouk_data['Area/Zip'].value_counts()[:10].values.tolist()
```

```python

readcouk_data['Area/Zip'].value_counts()[:10].plot(kind='pie',title='Top 10 Areas for Cleaners job by Read-Co-UK Job Portal',autopct=make_autopct(values_by_Area),figsize=(15,10))
```

```python
values_by_Role = readcouk_data['Role'].value_counts()[:10].values.tolist()
```

```python
readcouk_data['Role'].value_counts()[:10].plot(kind='pie',title='Top 10 Job Titles for Cleaners by Read.co.uk Job Portal',autopct=make_autopct(values_by_Role),figsize=(18,10))
```

```python
gd = readcouk_data['Role'].value_counts().head(10)
ax = sns.countplot(y="Role", data=readcouk_data,order=readcouk_data.Role.value_counts().iloc[:10].index).set_title("Top 10 Job Titles for Cleaners from Read.co.uk Data")
```

```python
values_by_company = readcouk_data['Company'].value_counts()[:10].values.tolist()
readcouk_data['Company'].value_counts()[:10].plot(kind='pie',title='Most Ads by any Company for Cleaners on Read.co.uk Job Portal',autopct=make_autopct(values_by_company),figsize=(18,10))
```
