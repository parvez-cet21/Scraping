## Import Necessary Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

### working with Webdriver to get source of Page

```python
options = webdriver.ChromeOptions()

options.add_argument("--incognito")
browser = webdriver.Chrome("C:\chromedriver\chromedriver.exe", options = options)
browser.get("https://jobtoday.com/gb/jobs-cleaning/london")
#sleep(5)
#browser.close()
```

### Beautifying the source Page

```python
src = browser.page_source
soup=BeautifulSoup(src)
```

#### Getting necessary elements from the page

```python
job = soup.find_all('h3', attrs = {'class' : 'jsx-739776546 JobCardLarge-role'})
city_zip = soup.find_all('div', attrs = {'class' : 'jsx-739776546 JobCardLarge-address'})
time = soup.find_all('div', attrs = {'class' : 'jsx-739776546 JobCardLarge-updateDate'})
```

```python
job_titles = []
job_loc = []
job_time = []
for i in range(len(job)):
    job_titles.append(job[i].text.strip())
    job_loc.append(city_zip[i].text.strip())
    job_time.append(time[i].text.strip())
    
```

#### Iterating over all pages of Site

```python
for i in range(2,89):
    browser.get("https://jobtoday.com/gb/jobs-cleaning/london?page="+str(i))
    src = browser.page_source
    soup=BeautifulSoup(src)
    job = soup.find_all('h3', attrs = {'class' : 'jsx-739776546 JobCardLarge-role'})
    time = soup.find_all('div', attrs = {'class' : 'jsx-739776546 JobCardLarge-updateDate'})
    for i in range(len(job)):
        job_titles.append(job[i].text.strip())
        job_loc.append(city_zip[i].text.strip())
        job_time.append(time[i].text.strip())
```

```python
print(job_titles)
print(job_time)
```

```python
df = pd.DataFrame(list(zip(job_titles,job_loc,job_time)), columns=['Role','City/Zip','Datetime_Posted'])
df
```
