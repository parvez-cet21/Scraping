## Import Necessary Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

### working with Webdriver to get source of Page

```python
options = webdriver.ChromeOptions()

browser = webdriver.Chrome("C:\chromedriver\chromedriver.exe", options = options)
browser.get("https://www.totaljobs.com/jobs/cleaner/in-london?radius=0")
```

### Beautifying the source Page

```python
src = browser.page_source
soup=BeautifulSoup(src)
```

#### Getting necessary elements from the page

```python
company = soup.find_all('div', attrs = {'class' : 'sc-AxheI cOZDZM'})
job = soup.find_all('h2', attrs = {'class' : 'sc-fzqARJ iyolKq'})
city_zip = soup.find_all('li', attrs = {'class' : 'sc-fznzOf fKAMhD'})
```

```python
job_title = []
company_name = []
area = []

for i in range(len(job)):
    job_title.append(job[i].text.strip())
    company_name.append(company[i].text.strip())
    area.append(city_zip[i].text.strip())
    
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--incognito")

browser = webdriver.Chrome("C:\chromedriver\chromedriver.exe", options = options)

browser.get("https://www.totaljobs.com/jobs/cleaner/in-london?radius=0&page=2")
src = browser.page_source
soup=BeautifulSoup(src)
company = soup.find_all('div', attrs = {'class' : 'sc-AxheI cOZDZM'})
job = soup.find_all('h2', attrs = {'class' : 'sc-fzqARJ iyolKq'})
city_zip = soup.find_all('li', attrs = {'class' : 'sc-fznzOf fKAMhD'})
for i in range(len(job)):
        job_title.append(job[i].text.strip())
        company_name.append(company[i].text.strip())
        area.append(city_zip[i].text.strip())
       
```

#### Iterating over all pages of Site

```python
df = pd.DataFrame(list(zip(company_name, job_title, area)), columns=['Company_Name','Role','Area/Zip'])
df
```
