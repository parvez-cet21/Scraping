```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
indeed_data = pd.read_csv('Indeed_Jobs.csv',index_col='Unnamed: 0')
indeed_data.head()
```

```python
indeed_data['City/Zip'].value_counts().head(10)
```

```python
indeed_data['Role'].value_counts().head(10)
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_autopct
```

```python
values_by_Area = indeed_data['City/Zip'].value_counts()[:10].values.tolist()
```

```python

indeed_data['City/Zip'].value_counts()[:10].plot(kind='pie',title='Top 10 Areas for Cleaners job by Indeed Job Portal',autopct=make_autopct(values_by_Area),figsize=(15,10))
```

```python
values_by_Role = indeed_data['Role'].value_counts()[:10].values.tolist()
```

```python
indeed_data['Role'].value_counts()[:10].plot(kind='pie',title='Top 10 Job Titles for Cleaners by Indeed Job Portal',autopct=make_autopct(values_by_Role),figsize=(18,10))
```

```python
gd = indeed_data['Role'].value_counts().head(10)
ax = sns.countplot(y="Role", data=indeed_data,order=indeed_data.Role.value_counts().iloc[:10].index).set_title("Top 10 Job Titles for Cleaners from Indeed Data")
```
