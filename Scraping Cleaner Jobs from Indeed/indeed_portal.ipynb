## Import Necessary Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

### working with Webdriver to get source of Page

```python
options = webdriver.ChromeOptions()

options.add_argument("--incognito")
browser = webdriver.Chrome("C:\chromedriver\chromedriver.exe", options = options)
browser.get("https://uk.indeed.com/jobs?q=Cleaner&l=London&radius=5")
```

### Beautifying the source Page

```python
src = browser.page_source
soup=BeautifulSoup(src)
```

#### Getting necessary elements from the page

```python
company = soup.find_all('span', attrs = {'class' : 'companyName'})
job = soup.find_all('h2', attrs = {'class' : ['jobTitle jobTitle-color-purple jobTitle-newJob','jobTitle jobTitle-color-purple']})
city_zip = soup.find_all('div', attrs = {'class' : 'companyLocation'})
time = soup.find_all('span', attrs = {'class' : 'date'})
```

```python
company
```

```python
company_name = []
job_titles = []
job_loc = []
job_time = []
for i in range(len(job)):
    company_name.append(company[i].text.strip())
    job_titles.append(job[i].text.strip())
    job_loc.append(city_zip[i].text.strip())
    job_time.append(time[i].text.strip())
    
```

#### Iterating over all pages of Site

```python
for i in range(10,800,10):
    browser.get("https://uk.indeed.com/jobs?q=Cleaner&l=London&radius=5&start="+str(i))
    src = browser.page_source
    soup=BeautifulSoup(src)
    company = soup.find_all('span', attrs = {'class' : 'companyName'})
    job = soup.find_all('h2', attrs = {'class' : ['jobTitle jobTitle-color-purple jobTitle-newJob','jobTitle jobTitle-color-purple']})
    city_zip = soup.find_all('div', attrs = {'class' : 'companyLocation'})
    time = soup.find_all('span', attrs = {'class' : 'date'})
    for i in range(len(job)):
        company_name.append(company[i].text.strip())
        job_titles.append(job[i].text.strip())
        job_loc.append(city_zip[i].text.strip())
        job_time.append(time[i].text.strip())
```

```python
df = pd.DataFrame(list(zip(company_name,job_titles,job_loc,job_time)), columns=['Comapny_Name','Role','City/Zip','Datetime_Posted'])
df
```

### Creating a CSV file of finding data

```python
df.to_csv('Indeed_Jobs.csv')
```
